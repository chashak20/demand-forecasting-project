```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

print("Libraries imported succesfully")
```

```python
data_path = r"C:\demand_Forecasting\data"

print("Data load ho raha hai")

sales_train = pd.read_csv(os.path.join(data_path, "sales_train_evaluation.csv"))
calendar = pd.read_csv(os.path.join(data_path, "calendar.csv"))
prices = pd.read_csv(os.path.join(data_path, "sell_prices.csv"))

print("sales_train shape :", sales_train.shape)
print("calendar shape :", calendar.shape)
print("prices shape :", prices.shape)
```

```python
print("======Dataset Info=======")
print(f'Total items : {sales_train["item_id"].nunique()}')
print(f'Total stores : {sales_train["store_id"].nunique()}')
print(f'Total days : {sales_train.shape[1] - 5}')
print(f'Categories : {sales_train["cat_id"].unique()}')
print(f'States : {sales_train["state_id"].unique()}')

```

```python
# Wide → Long format
print('🔄 Converting to long format (thoda time lagega)...')

id_cols   = ['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']
day_cols  = [c for c in sales_train.columns if c.startswith('d_')]

sales_long = sales_train.melt(
    id_vars    = id_cols,
    value_vars = day_cols,
    var_name   = 'd',
    value_name = 'sales'
)

# Calendar ke saath merge karo (date column lene ke liye)
sales_long = sales_long.merge(calendar[['d', 'date', 'wday', 'month', 'year', 'event_name_1']], 
                               on='d', how='left')
sales_long['date'] = pd.to_datetime(sales_long['date'])

print(f'✅ Long format shape: {sales_long.shape}')
sales_long.head()
```

```python
# Total daily sales
daily_sales = sales_long.groupby('date')['sales'].sum().reset_index()

fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Plot 1: Raw daily sales
axes[0].plot(daily_sales['date'], daily_sales['sales'], color='steelblue', linewidth=0.8, alpha=0.7)
axes[0].set_title('Total Daily Sales — Raw Trend', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Total Sales')

# Plot 2: 30-day rolling average (smooth trend)
daily_sales['rolling_30'] = daily_sales['sales'].rolling(30).mean()
axes[1].plot(daily_sales['date'], daily_sales['sales'],      color='lightblue',  linewidth=0.6, alpha=0.5, label='Daily')
axes[1].plot(daily_sales['date'], daily_sales['rolling_30'], color='darkblue',   linewidth=2,   label='30-day average')
axes[1].set_title('Total Daily Sales — 30-Day Rolling Average (Smooth Trend)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Sales')
axes[1].legend()

plt.tight_layout()
plt.savefig('trend_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 trend_analysis.png save ho gayi!')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Weekly pattern (wday: 1=Saturday ... 7=Friday in M5)
day_names   = {1:'Sat', 2:'Sun', 3:'Mon', 4:'Tue', 5:'Wed', 6:'Thu', 7:'Fri'}
weekly_avg  = sales_long.groupby('wday')['sales'].mean()
weekly_avg.index = [day_names[d] for d in weekly_avg.index]

axes[0].bar(weekly_avg.index, weekly_avg.values, color='steelblue', edgecolor='white')
axes[0].set_title('Average Sales by Day of Week', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Avg Sales')

# Monthly pattern
month_names  = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
                7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
monthly_avg  = sales_long.groupby('month')['sales'].mean()
monthly_avg.index = [month_names[m] for m in monthly_avg.index]

axes[1].bar(monthly_avg.index, monthly_avg.values, color='coral', edgecolor='white')
axes[1].set_title('Average Sales by Month', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Avg Sales')

plt.tight_layout()
plt.savefig('seasonality_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 seasonality_analysis.png save ho gayi!')
```

```python
# Events wale din mark karo
event_days   = sales_long[sales_long['event_name_1'].notna()].groupby(
                   ['date', 'event_name_1'])['sales'].sum().reset_index()
top_events   = event_days.groupby('event_name_1')['sales'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
top_events.plot(kind='barh', color='tomato', edgecolor='white')
plt.title('Top 10 Events by Average Sales Impact', fontsize=13, fontweight='bold')
plt.xlabel('Avg Sales on Event Day')
plt.tight_layout()
plt.savefig('event_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 event_analysis.png save ho gayi!')
```

```python
# Category wise daily sales
cat_daily = sales_long.groupby(['date', 'cat_id'])['sales'].sum().reset_index()

plt.figure(figsize=(16, 5))
for cat in cat_daily['cat_id'].unique():
    subset = cat_daily[cat_daily['cat_id'] == cat]
    plt.plot(subset['date'], subset['sales'], label=cat, linewidth=1)

plt.title('Daily Sales by Category — FOODS vs HOBBIES vs HOUSEHOLD', fontsize=13, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.savefig('category_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 category_analysis.png save ho gayi!')
```

```python
# Ek specific store ka data save karo modelling ke liye
# (CA_1 store, FOODS category — sab models pe test karenge isko)
store_data = sales_long[
    (sales_long['store_id'] == 'CA_1') & 
    (sales_long['cat_id']   == 'FOODS')
].groupby('date')['sales'].sum().reset_index()

store_data.to_csv('C:\demand_Forecasting\data.csv', index=False)

print(f'✅ CA_1 FOODS daily data saved!')
print(f'   Shape: {store_data.shape}')
print(f'   Date range: {store_data["date"].min()} → {store_data["date"].max()}')
store_data.tail()
```
