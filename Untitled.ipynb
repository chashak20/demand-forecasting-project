```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

print("Libraries imported succesfully")
```

```python
data_path = r"C:\demand_Forecasting\data"

print("Data load ho raha hai")

sales_train = pd.read_csv(os.path.join(data_path, "sales_train_evaluation.csv"))
calendar = pd.read_csv(os.path.join(data_path, "calendar.csv"))
prices = pd.read_csv(os.path.join(data_path, "sell_prices.csv"))

print("sales_train shape :", sales_train.shape)
print("calendar shape :", calendar.shape)
print("prices shape :", prices.shape)
```

```python
# Wide → Long format
print('🔄 Converting to long format (thoda time lagega)...')

id_cols   = ['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']
day_cols  = [c for c in sales_train.columns if c.startswith('d_')]

sales_long = sales_train.melt(
    id_vars    = id_cols,
    value_vars = day_cols,
    var_name   = 'd',
    value_name = 'sales'
)

# Calendar ke saath merge karo (date column lene ke liye)
sales_long = sales_long.merge(calendar[['d', 'date', 'wday', 'month', 'year', 'event_name_1']], 
                               on='d', how='left')
sales_long['date'] = pd.to_datetime(sales_long['date'])

print(f'✅ Long format shape: {sales_long.shape}')
sales_long.head()
```

```python
# Total daily sales
daily_sales = sales_long.groupby('date')['sales'].sum().reset_index()

fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Plot 1: Raw daily sales
axes[0].plot(daily_sales['date'], daily_sales['sales'], color='steelblue', linewidth=0.8, alpha=0.7)
axes[0].set_title('Total Daily Sales — Raw Trend', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Total Sales')

# Plot 2: 30-day rolling average (smooth trend)
daily_sales['rolling_30'] = daily_sales['sales'].rolling(30).mean()
axes[1].plot(daily_sales['date'], daily_sales['sales'],      color='lightblue',  linewidth=0.6, alpha=0.5, label='Daily')
axes[1].plot(daily_sales['date'], daily_sales['rolling_30'], color='darkblue',   linewidth=2,   label='30-day average')
axes[1].set_title('Total Daily Sales — 30-Day Rolling Average (Smooth Trend)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Sales')
axes[1].legend()

plt.tight_layout()
plt.savefig('trend_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 trend_analysis.png save ho gayi!')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Weekly pattern (wday: 1=Saturday ... 7=Friday in M5)
day_names   = {1:'Sat', 2:'Sun', 3:'Mon', 4:'Tue', 5:'Wed', 6:'Thu', 7:'Fri'}
weekly_avg  = sales_long.groupby('wday')['sales'].mean()
weekly_avg.index = [day_names[d] for d in weekly_avg.index]

axes[0].bar(weekly_avg.index, weekly_avg.values, color='steelblue', edgecolor='white')
axes[0].set_title('Average Sales by Day of Week', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Avg Sales')

# Monthly pattern
month_names  = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
                7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
monthly_avg  = sales_long.groupby('month')['sales'].mean()
monthly_avg.index = [month_names[m] for m in monthly_avg.index]

axes[1].bar(monthly_avg.index, monthly_avg.values, color='coral', edgecolor='white')
axes[1].set_title('Average Sales by Month', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Avg Sales')

plt.tight_layout()
plt.savefig('seasonality_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 seasonality_analysis.png save ho gayi!')
```

```python
# Events wale din mark karo
event_days   = sales_long[sales_long['event_name_1'].notna()].groupby(
                   ['date', 'event_name_1'])['sales'].sum().reset_index()
top_events   = event_days.groupby('event_name_1')['sales'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
top_events.plot(kind='barh', color='tomato', edgecolor='white')
plt.title('Top 10 Events by Average Sales Impact', fontsize=13, fontweight='bold')
plt.xlabel('Avg Sales on Event Day')
plt.tight_layout()
plt.savefig('event_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 event_analysis.png save ho gayi!')
```

```python
# Category wise daily sales
cat_daily = sales_long.groupby(['date', 'cat_id'])['sales'].sum().reset_index()

plt.figure(figsize=(16, 5))
for cat in cat_daily['cat_id'].unique():
    subset = cat_daily[cat_daily['cat_id'] == cat]
    plt.plot(subset['date'], subset['sales'], label=cat, linewidth=1)

plt.title('Daily Sales by Category — FOODS vs HOBBIES vs HOUSEHOLD', fontsize=13, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.savefig('category_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 category_analysis.png save ho gayi!')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

# ARIMA ke liye
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Metrics ke liye
from sklearn.metrics import mean_absolute_error, mean_squared_error

plt.style.use('seaborn-v0_8-whitegrid')

print('✅ Libraries imported!')
```

```python
# EDA notebook ne ye file save ki thi
# Agar file nahi mili toh DATA_PATH apna daalo
DATA_PATH = r"C:\demand_Forecasting\data" + "\\"

try:
    df = pd.read_csv(DATA_PATH + 'ca1_foods_daily.csv', parse_dates=['date'])
    print('✅ ca1_foods_daily.csv load ho gayi!')
except FileNotFoundError:
    # Agar EDA notebook nahi chali toh seedha banao
    print('⚠️  ca1_foods_daily.csv nahi mili — seedha bana raha hoon...')
    sales = pd.read_csv(DATA_PATH + 'sales_train_evaluation.csv')
    cal   = pd.read_csv(DATA_PATH + 'calendar.csv')
    
    id_cols  = ['id','item_id','dept_id','cat_id','store_id','state_id']
    day_cols = [c for c in sales.columns if c.startswith('d_')]
    
    long = sales[sales['store_id']=='CA_1'].melt(
        id_vars=id_cols, value_vars=day_cols, var_name='d', value_name='sales')
    long = long.merge(cal[['d','date']], on='d', how='left')
    long['date'] = pd.to_datetime(long['date'])
    
    df = long[long['cat_id']=='FOODS'].groupby('date')['sales'].sum().reset_index()
    df.to_csv(DATA_PATH + 'ca1_foods_daily.csv', index=False)
    print('✅ Data ready!')

df = df.set_index('date').asfreq('D')
print(f'Shape      : {df.shape}')
print(f'Date range : {df.index.min()} → {df.index.max()}')
df.head()
```

```python
plt.figure(figsize=(16, 4))
plt.plot(df.index, df['sales'], color='steelblue', linewidth=0.8)
plt.title('CA_1 Store — FOODS Daily Sales', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
def adf_test(series, name='Series'):
    result = adfuller(series.dropna())
    print(f'=== ADF Test: {name} ===')
    print(f'ADF Statistic : {result[0]:.4f}')
    print(f'p-value       : {result[1]:.4f}')
    if result[1] < 0.05:
        print('✅ Stationary hai! (p < 0.05) — ARIMA directly use kar sakte hain')
    else:
        print('⚠️  Non-stationary hai — differencing karni padegi (d=1 use karo)')
    print()

# Original series test
adf_test(df['sales'], 'Original Sales')

# Differenced series test
adf_test(df['sales'].diff(), 'Differenced Sales (d=1)')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 4))

plot_acf( df['sales'].diff().dropna(), lags=40, ax=axes[0],
          title='ACF — q parameter ke liye dekho')
plot_pacf(df['sales'].diff().dropna(), lags=40, ax=axes[1],
          title='PACF — p parameter ke liye dekho')

plt.tight_layout()
plt.savefig('acf_pacf.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 acf_pacf.png saved!')
```

```python
TEST_DAYS = 28

train = df.iloc[:-TEST_DAYS]
test  = df.iloc[-TEST_DAYS:]

print(f'Train size : {len(train)} days ({train.index.min().date()} → {train.index.max().date()})')
print(f'Test size  : {len(test)}  days ({test.index.min().date()} → {test.index.max().date()})')

plt.figure(figsize=(16, 4))
plt.plot(train.index, train['sales'], color='steelblue', label='Train')
plt.plot(test.index,  test['sales'],  color='orange',    label='Test (actual)')
plt.axvline(x=test.index[0], color='red', linestyle='--', label='Split point')
plt.title('Train / Test Split', fontsize=13, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print('🔄 ARIMA model train ho raha hai... (1-2 min lagega)')

arima_model = SARIMAX(
    train['sales'],
    order=(1, 1, 1),          # p=1, d=1, q=1
    enforce_stationarity=False,
    enforce_invertibility=False
)

arima_result = arima_model.fit(disp=False)
print('✅ ARIMA(1,1,1) trained!')
print(arima_result.summary())
```

```python
print('🔄 SARIMA model train ho raha hai... (2-3 min lagega)')

sarima_model = SARIMAX(
    train['sales'],
    order=(1, 1, 1),           # Non-seasonal
    seasonal_order=(1, 1, 1, 7),  # Seasonal — weekly (s=7)
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_result = sarima_model.fit(disp=False)
print('✅ SARIMA(1,1,1)(1,1,1,7) trained!')
```

```python
# ARIMA forecast
arima_forecast = arima_result.forecast(steps=TEST_DAYS)
arima_forecast.index = test.index

# SARIMA forecast
sarima_forecast = sarima_result.forecast(steps=TEST_DAYS)
sarima_forecast.index = test.index

print('✅ Forecasts ready!')
print(f'ARIMA  first prediction : {arima_forecast.iloc[0]:.1f}')
print(f'SARIMA first prediction : {sarima_forecast.iloc[0]:.1f}')
print(f'Actual first value      : {test["sales"].iloc[0]:.1f}')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Last 90 days of train + full test for context
for ax, forecast, title, color in zip(
    axes,
    [arima_forecast, sarima_forecast],
    ['ARIMA(1,1,1)', 'SARIMA(1,1,1)(1,1,1,7)'],
    ['tomato', 'green']
):
    ax.plot(train.iloc[-90:].index, train.iloc[-90:]['sales'],
            color='steelblue', label='Train (last 90 days)')
    ax.plot(test.index, test['sales'],
            color='orange', linewidth=2, label='Actual')
    ax.plot(test.index, forecast,
            color=color, linewidth=2, linestyle='--', label=f'{title} Forecast')
    ax.axvline(x=test.index[0], color='gray', linestyle=':', alpha=0.7)
    ax.set_title(f'{title} — Forecast vs Actual', fontsize=13, fontweight='bold')
    ax.legend()
    ax.set_ylabel('Sales')

plt.tight_layout()
plt.savefig('arima_sarima_forecast.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 arima_sarima_forecast.png saved!')
```

```python
actual = test['sales'].values

def evaluate(actual, predicted, model_name):
    mae  = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    print(f'📊 {model_name}')
    print(f'   MAE  : {mae:.2f}  (avg kitna galat hua)')
    print(f'   RMSE : {rmse:.2f} (bade errors ko zyada penalize karta hai)')
    print(f'   MAPE : {mape:.2f}% (percentage mein error)')
    print()
    return {'Model': model_name, 'MAE': round(mae,2), 'RMSE': round(rmse,2), 'MAPE': round(mape,2)}

results = []
results.append(evaluate(actual, arima_forecast.values,  'ARIMA(1,1,1)'))
results.append(evaluate(actual, sarima_forecast.values, 'SARIMA(1,1,1)(1,1,1,7)'))

results_df = pd.DataFrame(results)
print('=== MODEL COMPARISON TABLE ===')
results_df
```
