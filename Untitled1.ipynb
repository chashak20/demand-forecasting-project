```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Deep Learning ke liye
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Preprocessing ke liye
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

plt.style.use('seaborn-v0_8-whitegrid')

print('✅ Libraries imported!')
print(f'✅ TensorFlow version: {tf.__version__}')
```

```python
DATA_PATH = r"C:\demand_Forecasting\data" + "\\"

try:
    df = pd.read_csv(DATA_PATH + 'ca1_foods_daily.csv', parse_dates=['date'])
    print('✅ ca1_foods_daily.csv load ho gayi!')
except FileNotFoundError:
    print('⚠️  ca1_foods_daily.csv nahi mili — seedha bana raha hoon...')
    sales = pd.read_csv(DATA_PATH + 'sales_train_evaluation.csv')
    cal   = pd.read_csv(DATA_PATH + 'calendar.csv')
    id_cols  = ['id','item_id','dept_id','cat_id','store_id','state_id']
    day_cols = [c for c in sales.columns if c.startswith('d_')]
    long = sales[sales['store_id']=='CA_1'].melt(
        id_vars=id_cols, value_vars=day_cols, var_name='d', value_name='sales')
    long = long.merge(cal[['d','date']], on='d', how='left')
    long['date'] = pd.to_datetime(long['date'])
    df = long[long['cat_id']=='FOODS'].groupby('date')['sales'].sum().reset_index()
    df.to_csv(DATA_PATH + 'ca1_foods_daily.csv', index=False)
    print('✅ Data banaya!')

df = df.set_index('date').asfreq('D')
print(f'Shape      : {df.shape}')
print(f'Date range : {df.index.min()} → {df.index.max()}')
df.head()
```

```python
# Scale karo 0 to 1 range mein
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df[['sales']])

print(f'Original data range  : {df["sales"].min():.0f} → {df["sales"].max():.0f}')
print(f'Scaled data range    : {scaled_data.min():.4f} → {scaled_data.max():.4f}')
print('✅ Data scaled!')
```

```python
def create_sequences(data, lookback=30):
    """
    lookback = kitne past days dekh ke predict karna hai
    """
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, 0])   # past 'lookback' days
        y.append(data[i, 0])               # next day
    return np.array(X), np.array(y)

LOOKBACK  = 30    # past 30 days dekh ke predict karo
TEST_DAYS = 28

X_all, y_all = create_sequences(scaled_data, LOOKBACK)

# Train / Test split
X_train = X_all[:-TEST_DAYS]
y_train = y_all[:-TEST_DAYS]
X_test  = X_all[-TEST_DAYS:]
y_test  = y_all[-TEST_DAYS:]

# LSTM ke liye reshape: (samples, timesteps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test  = X_test.reshape(X_test.shape[0],   X_test.shape[1],  1)

print(f'X_train shape : {X_train.shape}  → (samples, timesteps, features)')
print(f'y_train shape : {y_train.shape}')
print(f'X_test shape  : {X_test.shape}')
print(f'y_test shape  : {y_test.shape}')
print('✅ Sequences ready!')
```

```python
def build_lstm_model(lookback):
    model = Sequential([
        # Layer 1: LSTM — 64 units, sequences return karo
        LSTM(64, return_sequences=True, input_shape=(lookback, 1)),
        Dropout(0.2),    # Overfitting rokne ke liye

        # Layer 2: LSTM — 32 units
        LSTM(32, return_sequences=False),
        Dropout(0.2),

        # Output Layer: 1 value predict karo
        Dense(16, activation='relu'),
        Dense(1)
    ])

    model.compile(
        optimizer='adam',
        loss='mean_squared_error'
    )
    return model

model = build_lstm_model(LOOKBACK)
model.summary()
```

```python
# EarlyStopping — agar improvement band ho jaaye toh ruk jao
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,          # 10 epochs tak improvement nahi → stop
    restore_best_weights=True
)

print('🔄 LSTM training shuru ho rahi hai...')
print('   (5-10 min lag sakta hai — chai pee lo ☕)')
print()

history = model.fit(
    X_train, y_train,
    epochs          = 100,
    batch_size      = 32,
    validation_split= 0.1,    # 10% validation ke liye
    callbacks       = [early_stop],
    verbose         = 1
)

print()
print(f'✅ Training complete!')
print(f'   Total epochs run : {len(history.history["loss"])}')
print(f'   Final train loss : {history.history["loss"][-1]:.6f}')
print(f'   Final val loss   : {history.history["val_loss"][-1]:.6f}')
```

```python
# Predict karo
pred_scaled = model.predict(X_test)

# Inverse transform — scaled values wapis original scale mein
pred_lstm = scaler.inverse_transform(pred_scaled).flatten()
actual    = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()

# Negative values 0 karo
pred_lstm = np.maximum(pred_lstm, 0)

print('✅ Predictions ready!')
print(f'First prediction : {pred_lstm[0]:.1f}')
print(f'Actual value     : {actual[0]:.1f}')
```

```python
# Test dates nikalo
test_dates = df.index[-TEST_DAYS:]
train_dates_plot = df.index[-TEST_DAYS-90:-TEST_DAYS]  # last 90 days of train
train_actual_plot = scaler.inverse_transform(
    scaled_data[-TEST_DAYS-90:-TEST_DAYS]).flatten()

plt.figure(figsize=(16, 5))
plt.plot(train_dates_plot, train_actual_plot,
         color='steelblue', label='Train (last 90 days)')
plt.plot(test_dates, actual,
         color='orange', linewidth=2, label='Actual')
plt.plot(test_dates, pred_lstm,
         color='red', linewidth=2, linestyle='--', label='LSTM Forecast')
plt.axvline(x=test_dates[0], color='gray', linestyle=':', alpha=0.7)
plt.title('LSTM — Forecast vs Actual (28 days)', fontsize=13, fontweight='bold')
plt.legend()
plt.ylabel('Sales')
plt.tight_layout()
plt.savefig('lstm_forecast.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 lstm_forecast.png saved!')
```

```python
mae  = mean_absolute_error(actual, pred_lstm)
rmse = np.sqrt(mean_squared_error(actual, pred_lstm))
mape = np.mean(np.abs((actual - pred_lstm) / actual)) * 100

lstm_result = {'Model': 'LSTM', 'MAE': round(mae,2),
               'RMSE': round(rmse,2), 'MAPE': round(mape,2)}

print('📊 LSTM Model')
print(f'   MAE  : {mae:.2f}')
print(f'   RMSE : {rmse:.2f}')
print(f'   MAPE : {mape:.2f}%')
```

```python
try:
    prev_results = pd.read_csv(DATA_PATH + 'model_comparison.csv')
    final_results = pd.concat(
        [prev_results, pd.DataFrame([lstm_result])], ignore_index=True)
except:
    final_results = pd.DataFrame([lstm_result])

print('=' * 55)
print('  🏆 FINAL MODEL COMPARISON — ALL MODELS')
print('=' * 55)
print(final_results.to_string(index=False))
print('=' * 55)

best = final_results.loc[final_results['MAE'].idxmin()]
print(f'\n🥇 Best overall model (by MAE): {best["Model"]}')
print(f'   MAE  = {best["MAE"]}')
print(f'   RMSE = {best["RMSE"]}')
print(f'   MAPE = {best["MAPE"]}%')
```

```python
colors = ['steelblue', 'coral', 'purple', 'darkgreen', 'red']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, metric in zip(axes, ['MAE', 'RMSE', 'MAPE']):
    bars = ax.bar(
        final_results['Model'],
        final_results[metric],
        color=colors[:len(final_results)],
        edgecolor='white', width=0.5
    )
    ax.set_title(f'{metric} — Lower is Better ↓', fontsize=12, fontweight='bold')
    ax.set_ylabel(metric)
    ax.tick_params(axis='x', rotation=20)
    # Best model gold border
    min_idx = final_results[metric].idxmin()
    bars[min_idx].set_edgecolor('gold')
    bars[min_idx].set_linewidth(3)

plt.suptitle('🏆 Final Model Comparison — ARIMA vs Prophet vs LSTM',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('final_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print('💾 final_model_comparison.png saved!')
print('(Gold border = winner for that metric)')
```
